```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
from celluloid import Camera
from IPython.display import HTML

# Create figure with reduced size
fig, ax = plt.subplots(figsize=(10, 12))
camera = Camera(fig)

# Set limits - reduced to focus on the monument
ax.set_xlim(-15, 15)
ax.set_ylim(0, 30)
ax.set_aspect('equal')
ax.set_facecolor('black')  # Black background

# Remove all grid lines and spines
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(axis='both', which='both', length=0)
ax.set_xticks([])
ax.set_yticks([])

# Title and labels (optional - you can remove these too if you want)
ax.set_title('Building the Central Shaheed Minar', fontsize=18, fontweight='bold', color='white')
ax.set_xlabel('Width (meters)', fontsize=10, color='white')
ax.set_ylabel('Height (meters)', fontsize=10, color='white')

# Colors
colors = {
    'base': '#5D5D5D',
    'pillar': '#E8E8E8',
    'top': '#A9A9A9',
    'railing': '#A9A9A9',
    'red': '#FF0000',
    'green': '#006400',
    'shadow': '#D3D3D3'
}

# Building steps
steps = 50

for step in range(steps + 1):
    progress = step / steps

    # Ground (always there)
    ground = Rectangle((-15, 0), 30, 1, color=colors['green'])
    ax.add_patch(ground)

    # Step 1: Base platforms
    if progress >= 0.1:
        platform1 = Rectangle((-12, 1), 24, 1.5, color=colors['base'],
                            alpha=min(1, (progress-0.05)*10))
        ax.add_patch(platform1)

    if progress >= 0.2:
        platform2 = Rectangle((-9, 2.5), 18, 1.5, color=colors['base'],
                            alpha=min(1, (progress-0.15)*10))
        ax.add_patch(platform2)

    if progress >= 0.3:
        platform3 = Rectangle((-6, 4), 12, 1.5, color=colors['base'],
                            alpha=min(1, (progress-0.25)*10))
        ax.add_patch(platform3)

    # Step 2: Main pillars (4 pillars)
    if progress >= 0.4:
        pillar_positions = [-5, -1.5, 1.5, 5]
        for i, pos in enumerate(pillar_positions):
            pillar_height = 14 if i in [1, 2] else 11
            pillar = Rectangle((pos - 0.8, 5.5), 1.6, pillar_height * min(1, (progress-0.35)*4),
                             color=colors['pillar'], alpha=0.9)
            ax.add_patch(pillar)

    # Step 3: Side pillars (additional 4 pillars)
    if progress >= 0.45:
        side_positions = [-9, -7, 7, 9]
        for pos in side_positions:
            side_pillar = Rectangle((pos - 0.8, 5.5), 1.6, 8 * min(1, (progress-0.4)*6),
                                  color=colors['pillar'], alpha=0.9)
            ax.add_patch(side_pillar)

    # Step 4: NO TOP LINES OR CONNECTORS - removed all line elements

    # Step 5: NO RAILINGS - removed all line elements

    # Step 6: Red sun - BIGGER (and only the circle, no rays)
    if progress >= 0.7:
        # Bigger red circle
        sun = Circle((0, 12), 3.5, color=colors['red'],  # Even bigger
                    alpha=min(0.9, (progress-0.65)*10))
        ax.add_patch(sun)

        # Add a slight glow effect
        sun_glow = Circle((0, 12), 3.9, color=colors['red'],
                         alpha=min(0.3, (progress-0.65)*8))
        ax.add_patch(sun_glow)

    # Step 7: Text and final details
    if progress >= 0.8:
        ax.text(0, 26, 'শহীদ মিনার', fontsize=18, fontname='Bengali',
                ha='center', color='white', fontweight='bold',
                alpha=min(1, (progress-0.75)*10))
        ax.text(0, 24, 'SHAHEED MINAR', fontsize=14, ha='center',
                color='white', fontweight='bold',
                alpha=min(1, (progress-0.75)*10))
        ax.text(0, 22, 'Est. 1952', fontsize=10, ha='center',
                color='white', alpha=min(0.8, (progress-0.75)*10))

    # Shadow effect
    if progress >= 0.3:
        shadow = Rectangle((-15, 0), 30, 0.5, color=colors['shadow'],
                          alpha=min(0.3, progress))
        ax.add_patch(shadow)

    camera.snap()

# Create animation
animation = camera.animate(interval=80, repeat=True)
HTML(animation.to_html5_video())
```
